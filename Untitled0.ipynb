```
import pandas as pd
import numpy as np

df = pd.read_csv("drive/My Drive/.Data/order_brush_order.csv").sort_values(by =['event_time'])
shopid = df.shopid.unique()
criminal = ["0"] * len(shopid)
for sid in shopid:
  temp = []
  cur = df[df.shopid.isin([sid])]
  for index,row in cur.iterrows():
    process = cur[(pd.to_datetime(cur.event_time) >= pd.to_datetime(row.event_time)) & (pd.to_datetime(cur.event_time) < pd.to_datetime(row.event_time) + pd.Timedelta(hours = 1,seconds = 1))].sort_values(by = ['userid'])
    process_orderid = process.orderid.unique()
    process_userid_unique = process.userid.unique()
    if len(process_orderid) / len(process_userid_unique) >= 3:
      index = np.where(shopid == row.shopid)[0][0]
      temp.append(process.userid.mode()[0])
  temp = np.unique(temp)
  uid = "0"
  if len(temp) != 0:
    uid = str(temp[0])
    for i in range(1,len(temp)):
      uid += "&"+str(temp[i])
  criminal[np.where(shopid == sid)[0][0]] = uid
  
data = {'shopid' : shopid,'userid':criminal}
ans = pd.DataFrame(data)
ans.to_csv(r'ans.csv',index = False,header = True)
```

```
import pandas as pd
import numpy as np
time_zero = pd.to_datetime("2019-12-27")
df = pd.read_csv("drive/My Drive/.Data/order_brush_order.csv").sort_values(by =['event_time'])
df['unix_time'] = pd.to_datetime(df['event_time']).apply(lambda x: int((x - time_zero).total_seconds()))
df['sus'] = False
shopid = df.shopid.unique()

ulist_ret = []

for sid in shopid:
  cur = df[df.shopid == sid]
  cur = cur.reset_index(drop = True)
  n = len(cur)
  for start in range(n):
    for end in range(start,n):
      if cur.loc[end,'unix_time'] - cur.loc[start,'unix_time'] > 3600:
        break
      
      if start == 0 or end == n-1 or cur.loc[end+1,'unix_time'] - cur.loc[start-1,'unix_time'] > 3600:
        if end - start + 1 >= 3*cur.loc[start:end,'userid'].nunique():
          cur.loc[start:end,'sus'] = True
  cur = cur[cur.sus]
  udict = cur.userid.value_counts().to_dict()
  max_count = 0
  ulist = []
  for k,v in sorted(udict.items()):
    if v > max_count:
      max_count = v
      ulist = [k]
    elif v == max_count:
      ulist.append(k)
  rep = ''
  for s in ulist:
    if rep == '':
      rep = str(s)
    else:
      rep += '&' + str(s)
  if rep =='':
    rep = '0'
  ulist_ret.append(rep)

ans = pd.DataFrame({'shopid' : shopid,'userid' : ulist_ret})
ans.to_csv(r'submit.csv',index = False,header = True)
```
